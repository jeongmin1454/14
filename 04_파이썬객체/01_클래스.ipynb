```python
# 객체에 연산할 숫자 지정하기

class Fourcal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
cal6 = Fourcal()
print(type(cal6))
cal6.setdata(4, 2)

print(cal6.first)
print(cal6.second)
print(cal6.add())
```

```python
class Fourcal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
cal7 = Fourcal()

cal7.setdata(5, 6)
print(cal7.add())
print(cal7.sub())

```

```python
class Fourcal():
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
cal9 = Fourcal(3, 4)
print(cal9.add())
print(cal9.sub())
print(cal9.mul())
print(cal9.div())
```
