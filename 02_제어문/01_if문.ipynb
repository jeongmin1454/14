```python
# if문은 왜 필요할까?
# 돈아 있으면 택시를 타고 가고, 돈이 없으면 걸어간다.


money = True
if money:
    print("택시를 타고 간다")

else:
    print("걸어가라")
```

```python
# 돈이 30000원 이상 있거나 카드가 있다면 택시를 타고 가고
# 그렇지 않으면 걸어가라
money = 30000
card = True

if money>=30000 or card == True:
    print("택시")
else:
    print("걸어가라")
```

```python
# 만약에 주머니에 돈(money)이 있으면 택시를 타고 가고
# 없으면 걸어가라

pocket =["paper", "phone", "money"]

if "money" in pocket:
    print("택시")
else:
    print("걸어가라")
```

```python
# 다양한 조건을 판단하는 elif

# 주머니에 돈이 있으면 택시를 타고 가고
# 주머니에 돈이 없지만 카드가 있으면 택시를 타고가고
# 돈도 없고 카드도 없으면 걸어가라

pocket =["paper", "phone", "money"]
card = True

if "money" in pocket:
    print("택시")

else:
    if card == True:
        print("택시")
    else:
        print("걸어가라")
```

```python
# elif

if "money" in pocket:
    print("택시")
elif card == True:
    print("택시")
else:
    print("걸어가라")
```

```python
# match-case 구문

# match 변수
#   case 패턴1:
#       수행할 문장
#   case 패턴2:
#       수행할 문장
```

```python
grade = input("A ~ F중 하나의 값을 입력하시오")

match grade:
    case "A" | "B" | "C" :
        print("합격입니다")
    case _:
        print("불합격입니다")
```
