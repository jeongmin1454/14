```python
# 변수를 만드는 여러가지 방법
a, b = ["python", "like"]       #개수 순서 같아야함
print(a)
print(b)
print(type(a))
```

```python
(a, b) = "python", "like"       
print(a)
print(b)
print(type(a))
```
