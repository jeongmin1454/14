```python
print(0, bool(0))      #0 빼고 다 트루
print(1, bool(1))
print(2, bool(2))
print("     ", bool(""))       #뭐라도 적으면 트루
print("python", bool("pyhton"))
print("123123", bool("123123"))
print("[]", bool([]))          #뭐라도 적으면 트루
print("[1, 2, 3]", bool([1, 2, 3]))

```

```python
# 논리연산자
# 1. and 연산자
print(True and True)
print(True and False)
print(False and False)
print(True and True and False)


print("=========================")
# 2. or 연산자
print(True or True)
print(True or False)
print(False or False)
print(False or False or True)


print("=========================")
# 3. not 연산자
print(not True)
print(not False)
print(not 1)
print(not[])
```
