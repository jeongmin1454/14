```python
def calculate(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다"
        return a / b
    else:
        return "사칙연산이 아닙니다"

calculate(3, 3, "*")
calculate(5, 0, "/")
```

```python
def get_average(*args):
    if len(args) == 0:
        return 0
    total = sum(args)
    avg = total / len(args)

    return round(avg, 2)

print(get_average(90, 88, 85))
```

```python
def prin_gugudan(dan, start=1, end=9):
    for i in range(start, end+1):
        result = dan * i
        print(f"{dan} X {i} = {result}")

    if start > end:
        print("XXXXXXXXXXXX")

prin_gugudan(3,1,9)
prin_gugudan(3,1,9)
```

```python
def left_triangle(height):
    for i in range(1, height+1):
        print("*"*i)

left_triangle(8)
```

```python
def check_even_odd(num):
    if num % 2 == 0:
        return("짝수")
    else:
        return("홀수")
   

print(check_even_odd(10))
print(check_even_odd(117))
```

```python
def my_abs(num):
    if num > 0:
        return(num)
    else:
        return(num*(-1))

print(my_abs(5))
print(my_abs(-3))
print(my_abs(0))
```

```python
def find_max(a, b ,c):

    if a > b and a > c:
        return(a)
    elif b > a and b > c:
        return(b)
    elif c > a and c > b:
        return(c)
    else:
        return("X")

    
print(find_max(62,16,2))   
```

```python
def repeat_name(name, count):
    for i in range(count):
        print(f"{name}")

repeat_name("정민", 5)
```

```python
def print_rectangle(width, height):
    for i in range(1, width+1):
        for j in range(1, height+1):
            print("*",end="")
        print()

print_rectangle(5, 6)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result

factorial(6)
```

```python
def list_sum(numbers):
    total = 0
    for num in numbers:
        total = total + num
        
    return total

print(list_sum([5, 6, 7]))
```

```python
def get_even_numbers(numbers):
    even = []
    for i in numbers:
        if i % 2 == 0:
            even.append(i)
    return even

get_even_numbers([1,2,3,4,5,6,7,8])
```

```python
def my_len(text):
    count = 0

    for i in text:
        count += 1

    return count

my_len("123456789")
```

```python
def count_char(text, ch):
    count = 0

    for char in text:
        if char == ch:
            count += 1

    return count

count_char("abcabcaaa", "a")
```

```python
def reverse_text(text):
    reverse = ""

    for char in text:
        reverse = char + reverse

    return reverse

reverse_text("123456")
```
