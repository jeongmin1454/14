```python
# 5번

id = "admin" 
password = 1234
chance = 3

while chance > 0:
    id_input = input("아이디를 입력하세요")
    password_input = int(input("비밀번호를 입력하세요"))

    if id_input == id and password_input == password:
        print("환영힙니다")
        break

    else:
        chance -= 1
        if chance > 0:
            print(f"실패 기횜 {chance}번 남음")
        else:
            print("계정 잠김")
```

```python
num = int(input("숫자를 입력하세요: "))
if num < 2:
    print("2 이상의 숫자만 입력 가능합니다.")
else:
    for i in range(2, num):
        if num % i == 0:
            print(f"{num}은(는) 소수가 아닙니다.")
            break
        else:
            print(f"{num}은(는) 소수입니다.")
            break

primes = []

for n in range(2, num + 1):
    prime = True
    for i in range(2, n):
        if n % i ==0:
            prime = False
            break
        if prime:
            primes.append(n)

print(f"1qnxj {num}까지의 소수:")
print(primes)
```

```python
money= 10000
while True:
    print("== ATM 시스템 ===")
    print("1. 잔액 조회") 
    print("2. 입금") 
    print("3. 출금") 
    print("4. 종료")

    option = input("선택하세요")

    if option == "1":
        print(f"현재 잔액: {money:,}원")
    elif option == "2":
        money_plus = int(input("입금액: "))
        money += money_plus
        print(f"입금 완료! 현재 잔액: {money:,}원")
    elif option == "3":
        money_min = int(input("출금액: "))
        if money_min > money:
            print(f"잔액이 부족합니다. 현재 잔액: {money:,}원")
        else:
            money -= money_min
            print(f"출금 완료! 현재 잔액: {money:,}원")
    elif option == "4":
        print("이용해 주셔서 감사합니다.")
        break
```

```python
import random

num = int(random.random() * 100)
chance = 1
print("1부터 100사이의 숫자를 맞춰보세요!")

while True:
    num_user = int(input(f"시도 {chance}: "))
    if num_user > num:
        print("Down")
        
    elif num_user < num:
        print("Up!")
        
    else:
        print(f"정답입니다! {chance}번 만에 맞추셨습니다!")
        break

    chance += 1
```

```python
year = int(input("년도: "))
month = int(input("월: "))

if month < 1 or month > 12:
    print("1~12월만 입력 가능합니다.")
else:
    if month in [1, 3, 5, 7, 8, 10, 12]:
        days = 31
    elif month in [4, 6, 9, 11]:
        days = 30
    else:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            days = 29
        else:
            days = 28

    print(f"\n    {year}년 {month}월")
    for day in range(1, days + 1):
        print(f"{day:2}", end=" ")
        if day % 7 == 0:
            print()
    print()
```

```python
student_num = int(input("학생 수를 입력하세요: "))
students = []

total_kor = 0
total_eng = 0
total_math = 0

for i in range(1, student_num + 1):
    print(f"{i}번 학생")
    name = input("이름: ")
    kor = int(input("국어: "))
    eng = int(input("영어: "))
    math = int(input("수학: "))
    average = (kor + eng + math) / 3
    if average >= 90:
        print(f"평균: {average}점 (등급: A)")
    elif average >= 80:
        print(f"평균: {average}점 (등급: B)")
    elif average >= 70:
        print(f"평균: {average}점 (등급: C)")
    elif average>= 60:
        print(f"평균: {average}점 (등급: D)")
    else:
        print(f"평균: {average}점 (등급: F)")

    students.append({"이름": name, "평균": average})
    total_kor = total_kor + kor
    total_eng = total_eng + eng
    total_math = total_math + math
print(f"국어 평균: {total_kor / student_num}점")
print(f"영어 평균: {total_eng / student_num}점")
print(f"수학 평균: {total_math / student_num}점")
```

```python
products = {"커피": 10000, "계란": 11000, "사과": 12000, "빵": 13000, "휴대폰": 1000000}
cart= []

while True:
    print("1.상품목록 2.장바구니담기 3.장바구니보기 4.결제 5.종료")
    option = input("원하는 메뉴 번호를 입력하세요: ")

    if option == "1":
        print("상품목록")
        for item, price in products.items():
            print(f"{item}:{price}원")
    elif option == "2":
        cart_in = input("상품을 선택하세요")
        if cart_in in products:
            cart.append(cart_in)
            print(f"{cart_in}을 담았습니다")
            print(f"현재 장바구니는{cart}입니다")
        else:
            print("상품이 없습니다")
    elif option == "3":
        print("장바구니 보기")
        if len(cart) == 0:
            print("비었습니다")
        else: 
            total = 0
            for product in cart:
                price = products[product] 
                print(f"- {product} : {price}원")
                total = total + price  
            print("현재 총 합계:", total, "원")

    elif option == "4":
        print("결제 진행 ")
        if len(cart) == 0:
            print("결제할 상품이 없습니다.")
        else:
            total = 0
            for product in cart:
                total = total + products[product]
            coupon = input("할인 쿠폰을 사용하시겠습니까? (yes/no): ")
            if coupon == "yes":
                total = int(total * 0.9)
                print("10% 할인이 적용되었습니다.")
            if total >= 30000:
                print("3만원 이상 구매로 배송비가 무료입니다")
            else:
                total = total + 3000
                print("배송비 3,000원이 추가되었습니다.")

            print(f"최종 결제 금액: {total}원")
            print("결제가 완료되었습니다. 장바구니를 비웁니다.")
            cart = [] 

    elif option == "5":
        print("프로그램을 종료합니다. 이용해 주셔서 감사합니다!")
        break

    else:
        print("잘못된 번호입니다. 다시 입력해 주세요.")
```
